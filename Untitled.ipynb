```python
import numpy as np
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
```

```python
db=sqlite3.connect("test.db")
url='E:\csv1\main\maindatabase%s.csv'
for i in [1,2,3,4]:
    print(url % i)
    chunks=pd.read_csv(url %i , chunksize=100_000)
    for chunk in chunks:
        chunk.columns=[column.replace(' ','_') for column in chunk.columns]
        chunk.to_sql('tripdata',db,if_exists='append')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_y_trans = le.fit_transform(data_y)
data_y_trans
```

```python
le_fid = LabelEncoder()
le_fid.fit(data_X['Flow_ID'])
data_X['Flow_ID'] = le_fid.fit_transform(data_X['Flow_ID'])

le_SIP = LabelEncoder()
le_SIP.fit(data_X['Src_IP'])
data_X['Src_IP'] = le_SIP.fit_transform(data_X['Src_IP'])

le_DIP = LabelEncoder()
le_DIP.fit(data_X['Dst_IP'])
data_X['Dst_IP'] = le_DIP.fit_transform(data_X['Dst_IP'])


le_flowbit = LabelEncoder()
le_flowbit.fit(data_X['Flow_Byts/s'])
data_X['Flow_Byts/s'] = le_flowbit.fit_transform(data_X['Flow_Byts/s'])

le_flowpk = LabelEncoder()
le_flowpk.fit(data_X['Flow_Pkts/s'])
data_X['Flow_Pkts/s'] = le_flowpk.fit_transform(data_X['Flow_Pkts/s'])

le_subcat = LabelEncoder()
le_subcat.fit(data_X['Sub_Cat'])
data_X['Sub_Cat'] = le_subcat.fit_transform(data_X['Sub_Cat'])

le_timestamp = LabelEncoder()
le_timestamp.fit(data_X['Timestamp'])
data_X['Timestamp'] = le_timestamp.fit_transform(data_X['Timestamp'])
data_X
```

```python
data_X.to_csv('maindataset.csv')
```

```python
from sklearn.feature_selection import chi2 
from sklearn.feature_selection import SelectKBest 
from sklearn.ensemble import ExtraTreesClassifier

#selecting 20 best features
# select_best= SelectKBest(chi2, k=20)
# X_feat_20 = select_best.fit_transform(data_X, data_y_trans)
# X_feat_20.shape

model = ExtraTreesClassifier(random_state=42)
model.fit(data_X, data_y_trans)
```

```python
model.feature_importances_
```

```python
feature_importance_std = pd.Series(model.feature_importances_, index=data_X.columns)
feature_importance_std.nlargest(20).plot(kind='bar', title='Standardised Dataset Feature Selection using ExtraTreesClassifier')
```
